```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Line Graph

```python
gas  = pd.read_csv('gas_prices.csv')
```

```python
print(gas)
```

```python
gas
```

```python
plt.plot(gas.Year, gas.USA)

plt.show()
```

```python
plt.plot(gas.Year, gas.Germany)

plt.show()
```

```python
plt.title('Varying Gas Prices with Countries')

plt.plot(gas.Year, gas.Germany, label='Germany')

plt.plot(gas.Year, gas.USA, label='USA')

plt.xlabel('Year')
plt.ylabel('USD')

plt.legend()

plt.show()
```

```python
plt.title('Varying Gas Prices with Countries')

plt.plot(gas.Year, gas.Germany, label='Germany')
plt.plot(gas.Year, gas['South Korea'], label='South Korea')
plt.plot(gas.Year, gas.USA, label='USA')

plt.xlabel('Year')
plt.ylabel('USD')

plt.legend()

plt.show()
```

```python
plt.title('Varying Gas Prices with Countries')

plt.plot(gas.Year, gas.Germany, label='Germany')
plt.plot(gas.Year, gas['South Korea'], label='South Korea')
plt.plot(gas.Year, gas.USA, label='USA')

plt.xlabel('Year')
plt.ylabel('USD')

plt.xticks(gas.Year[::2])

plt.legend()

plt.show()
```

```python
plt.title('Varying Gas Prices with Countries')

plt.plot(gas.Year, gas.Germany, 'b.--', label='Germany')
plt.plot(gas.Year, gas['South Korea'], 'r.--', label='South Korea' )
plt.plot(gas.Year, gas.USA, 'g.--', label='USA' )

plt.xlabel('Year')
plt.ylabel('USD')

plt.xticks(gas.Year[::2])

plt.legend()

plt.show()
```

```python
plt.title('Varying Gas Prices with Countries')

for country in gas:
    if country !='Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)

print(gas.Year[::3])

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('USD')

plt.legend()

plt.savefig('Gas_Prices_Figure.png', dpi=300)

plt.show()

```
