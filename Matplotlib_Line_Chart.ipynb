### Basic Graph


```python
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y)
```

plt.plot() plots a basic graph between given two value sets

it also prints the line : [<matplotlib.lines.Line2D at 0x196b9454370>]

```python
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y)

plt.show()
```

to remove the line : [<matplotlib.lines.Line2D at 0x196b9454370>]

we use plt.show() after plt.plot()
    
    

```python
x=[1,2,3]
y=[2,4,6]


plt.plot(x,y)

plt.title('Apun ka pahila graph bhidu!')

plt.show()
```

plt.title() : gives title to the graph

```python
x=[1,2,3]
y=[2,4,6]


plt.plot(x,y)

plt.title('Apun ka pahila graph bhidu!')

plt.xlabel('X Axis')

plt.ylabel('Y Axis')

plt.show()
```

plt.xlabel()

and 

plt.ylabel()

Gives labels (names) to x and y axes respectively.

you can also change the font by using fontdict

plt.title('Babu!', fontdict = { 'fontname' : 'Cosmic Sans MS' , 'fontsize' : 20})

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y)

plt.title("BABUMOSHAY!!")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.show()
```

if you want your values, then use xticks and yticks

plt.xticks(args)
plt.yticks(args)

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y)

plt.title("BABUMOSHAY!!")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])

plt.yticks([0,2,4,6,8])

plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y, label='2x')

plt.title("BABUMOSHAY!!")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])

plt.yticks([0,2,4,6,8])

plt.legend()

plt.show()
```

plt.legend()  allows to show the entered relationship between values at axes.  First we need to enter 'label' in plt.plot() section, then enter plt.legend()

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

plt.title("BABUMOSHAY!!")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])

plt.yticks([0,2,4,6,8])

plt.legend()

plt.show()
```

you can change line color, style, add markers, change marker style and color using above functions used inside plt.plot()

YOU CAN ALSO USE SHORTHAND NOTATION : plt.plot(x,y, '[line_color][marker_style][line_style]', label=' ') as:


```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y, 'r*--', label='2x' )

plt.title("BABUMOSHAY!!")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])

plt.yticks([0,2,4,6,8])

plt.legend()

plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y, 'b*--', label='2x' )

plt.title("BABUMOSHAY!!")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

## Line Number Two

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5], x2[:5]**2, 'r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])


plt.legend()

plt.show()
```

ADDING A SECOND LINE AS PER ABOVE

RESIZING THE GRAPH: plt.figure(figsize=(x,y), dpi=100)

SAVING THE GRAPH: plt.savefig('graph.png', dpi=300)

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y, 'b*--', label='2x' )

plt.title("BABUMOSHAY!!")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

## Line Number Two

# select interval for plotting points: here, points will be plotted at 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5
x2 = np.arange(0,4.5,0.5)

# plot part of graph as  li
plt.plot(x2[:5], x2[:5]**2, 'r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])


plt.legend()

plt.savefig('graph.png', dpi=300)

plt.show()
```
