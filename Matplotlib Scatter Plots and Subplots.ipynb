```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df
```

```python
df.head(5)
```

### Scatter Plot

```python
plt.scatter(df['sqft_living'], df['price'])
plt.title('Bedrooms v/s Price')
plt.xlabel('sq. ft')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(df['sqft_living'], df['price'], 1)
plt.title('Bedrooms v/s Price')
plt.xlabel('sq. ft')
plt.ylabel('Price')
plt.show()
```

here in plt.scatter(), 1 indicates size of the dots

```python
x = df.sqft_living
```

```python
line = x*300
line
```

```python
plt.plot(x,line,'r')
plt.xlim(0,16500)
plt.show()
```

### Subplots

```python
fig, babu = plt.subplots()
babu.plot(x, line, 'r', label='Prediction')
babu.scatter(df['sqft_living'], df.price, label='Training Data')
babu.legend(loc=2)
babu.set_xlabel('Size')
babu.set_ylabel('Price')
babu.set_title('Predicted price vs Size')
```

```python
plt.subplot(1,2,1)
#          1 row, 2 columns, 1st one
plt.plot(x,line)
plt.ylabel('Price')
plt.xlabel('Sq.ft')
plt.title('Line')

plt.subplot(1,2,2)
plt.scatter(df['sqft_living'], df.price/df.price.max())
plt.ylabel('Price')
plt.xlabel('Sq.ft')
plt.title('Scatter plot')

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,line)
plt.ylabel('Price')
plt.xlabel('Sq.ft')
plt.title('Line')

plt.subplot(2,1,2)
plt.scatter(df['sqft_living'], df.price/df.price.max())
plt.ylabel('Price')
plt.xlabel('Sq.ft')
plt.title('Scatter plot')

plt.show()
```
