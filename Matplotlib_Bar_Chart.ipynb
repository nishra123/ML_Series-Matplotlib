```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Bar chart

```python
labels = ['Shona', 'Bunty', 'Babu']
values = [1, 3, 2]

# nehmi x,y  => x pahile 
plt.bar(labels, values)

plt.show()
```

you can create a line chart using plt.plot() while you can create a bar chart using plt.bar()

remember: x first and y later!


```python
labels = ['Shona', 'Bunty', 'Babu']
values = [1, 3, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')

plt.show()
```

element.set_hatch('char')  shades that element with that char in bar chart

```python
labels = ['Shona', 'Bunty', 'Babu']
values = [1, 3, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('.')
bars[2].set_hatch('*')

plt.show()
```
