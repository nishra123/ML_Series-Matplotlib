```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Box Plot


for box plot we use plt.boxplot([Attribute1, Attribute2....])

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
barcelona
```

```python
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
madrid
```

```python
plt.boxplot([barcelona, madrid])

plt.show()
```

```python
labels = ['FC Barcelona', 'Real Madrid']

plt.boxplot([barcelona, madrid], labels=labels)

plt.show()
```

```python
plt.title("Soccer teams Comparison")

labels = ['FC Barcelona', 'Real Madrid']

plt.boxplot([barcelona, madrid], labels=labels)

plt.show()
```

```python
plt.title("Soccer teams Comparison")

labels = ['FC Barcelona', 'Real Madrid']
boxes = plt.boxplot([barcelona, madrid], labels=labels, patch_artist=True)

for box in boxes['boxes']:
    # set edge color
    box.set(color='b', linewidth=2)
    
    # change fill color
    box.set(facecolor='white')
    

plt.ylabel('FIFA Overall rating')
    
    
plt.show()
```
